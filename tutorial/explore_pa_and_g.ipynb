```python
import pickle
import pandas as pd
%matplotlib inline
```

```python
pitcher_filename = '04_oct_cleaned_pickled_pitcher_data.pkl'
batter_filename = '04_oct_cleaned_pickled_batter_data.pkl'
```

```python
with open(pitcher_filename, 'rb') as f_obj:
    p = pd.DataFrame(pickle.load(f_obj))
p = p.dropna()
```
